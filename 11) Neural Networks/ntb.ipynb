## Neural networks

### The Neuron

Neuron usually gets some input signals with different values (or potencials). Core of the neuron gets all the input signals from synapsis and process them to an output value that goes down through the axon. Every input has its `weight` or importance. It's important to notice that not every combination of inputs fires the neuron action (it depends on the shape of the `activation function` of the neuron). Every neuron has its own threshold, sometimes called `Bias`.

### The Activation Function

This function takes the sum of all weighted inputs as an input and returns appropriate output value. There are several different activation function as can be seen below.

- Threshold function
- Sigmoid
- Rectifier
- Tanh

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)

theta = np.heaviside(x, 0)
sigmoid = 1/(1 + np.exp(-x)) 
def relu(x):
    return np.maximum(0, x)
tanh = np.tanh(x)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
fig.suptitle('Different Activation Functions')
fig.set_size_inches(25,5)

ax1.plot(x, theta)
ax1.set_title("Threshold function")

ax2.plot(x, sigmoid)
ax2.set_title("Sigmoid function")

ax3.plot(x, relu(x))
ax3.set_title("Rectifier function")

ax4.plot(x, tanh)
ax4.set_title("Tanh(x) function")
```

### Gradient Descent

In the previous example we talked about updating the neuron weights and bias, the question is, how exactly we should update these parameters? Should we increase the values or decrease or increase just few of them? Those all are serious questions. If we would do it in some random way it could take aproximately $10^{50}$years to find the optimal values of all parameters. That means that we need some smarter way how to update those.<br>
<br>

As first step we should find the `gradient of the Cost function` with respect to the weights $\nabla_w (C)$ and to the bias $\nabla_b (C)$.

$$ \nabla_w (C) = (\frac{\partial C}{\partial w_1}, \frac{\partial C}{\partial w_2}, \dots, \frac{\partial C}{\partial w_n}) \qquad \wedge \qquad \nabla_b (C) = \frac{\partial C}{\partial b}  $$

Now we will update the weights and bias in the following manner:

$$ w_{new} = w_{old} + \gamma\nabla_w(C) \qquad \wedge \qquad b_{new} = b_{old} + \gamma\nabla_b(C), $$

where $\gamma$ is the learning rate. If we choose the right value of the learning rate, then we can find the optimal value of parameters in few iterations or in few Epochs.
