## Polynomial regression



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

### Linear, Quadratic, Cubic, etc. noisy data

```python
X = 10 * (np.random.rand(1000, 1) - 0.5)

linear = 4 + 110 * X + 50 * np.random.randn(1000, 1)
quadratic = 4 + 3 * X + 18 * X**2 + 50 * np.random.randn(1000, 1)
cubic = 4 + 3 * X + 8 * X**2 - 4 * X**3 + 50 * np.random.randn(1000, 1)
quattro = 4 + X + 2 * X**2 - X**3 + X**4 + 50 * np.random.randn(1000, 1)

fig, axes = plt.subplots(1, 4, figsize =(30, 6), sharey=True)
fig.suptitle('Polynomial noisy data')
axes[0].scatter(X, linear)
axes[1].scatter(X, quadratic)
axes[2].scatter(X, cubic)
axes[3].scatter(X, quattro)
plt.show()
```

### Linear regression line

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

X_train, X_test, y_train, y_test = train_test_split(X, linear, test_size=0.05, random_state=1234)

poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)

linear_model = LinearRegression()
linear_model.fit(X_train_poly, y_train)
coefs, bias = np.round(linear_model.coef_[0]), np.round(linear_model.intercept_[0])

x_axis = np.arange(-5, 5, 0.1)
line = bias + coefs[1] * x_axis + coefs[2] * x_axis**2

plt.scatter(X_test, y_test)
plt.plot(x_axis, line, c="r", label=f"y = {coefs[2]} * X^2 + {coefs[1]} * X + {bias}")
plt.legend()
```

### Quadratic regression line

```python
X_train, X_test, y_train, y_test = train_test_split(X, quadratic, test_size=0.05, random_state=1234)

poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)

quadratic_model = LinearRegression()
quadratic_model.fit(X_train_poly, y_train)
coefs, bias = np.round(quadratic_model.coef_[0]), np.round(quadratic_model.intercept_[0])

quadratic_pred = quadratic_model.predict(X_test_poly)
x_axis = np.arange(-5, 5, 0.1)
line = bias + coefs[1] * x_axis + coefs[2] * x_axis**2

plt.scatter(X_test, y_test)
plt.plot(x_axis, line, c="r", label=f"y = {coefs[2]} * X^2 + {coefs[1]} * X + {bias}")
plt.legend()
```

### Cubic regression line

```python
X_train, X_test, y_train, y_test = train_test_split(X, cubic, test_size=0.05, random_state=1234)

poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)

cubic_model = LinearRegression()
cubic_model.fit(X_train_poly, y_train)
coefs, bias = np.round(cubic_model.coef_[0]), np.round(cubic_model.intercept_[0])

x_axis = np.arange(-5, 5, 0.1)
line = bias + coefs[1] * x_axis + coefs[2] * x_axis**2 + coefs[3] * x_axis**3

plt.scatter(X_test, y_test)
plt.plot(x_axis, line, c="r", label=f"y = {coefs[3]} * X^3 + {coefs[2]} * X^2 + {coefs[1]} * X + {bias}")
plt.legend()
```

### Quattro regression line

```python
X_train, X_test, y_train, y_test = train_test_split(X, quattro, test_size=0.05, random_state=1234)

poly = PolynomialFeatures(degree=4)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)

quattro_model = LinearRegression()
quattro_model.fit(X_train_poly, y_train)
coefs, bias = np.round(quattro_model.coef_[0]), np.round(quattro_model.intercept_[0])

x_axis = np.arange(-5, 5, 0.1)
line = bias + coefs[1] * x_axis + coefs[2] * x_axis**2 + coefs[3] * x_axis**3 + coefs[4] * x_axis**4

plt.scatter(X_test, y_test)
plt.plot(x_axis, line, c="r", label=f"y = {coefs[4]} * X^4 + {coefs[3]} * X^3 + {coefs[2]} * X^2 + {coefs[1]} * X + {bias}")
plt.legend()
```
